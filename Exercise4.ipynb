```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
from scipy import optimize


exp_data = np.load("I_q_IPA_exp.npy")
th_data = np.load("I_q_IPA_model.npy")

f_th = interpolate.interp1d(th_data[:,0], th_data[:,1])
#x_th = np.linspace(0,30,100)
#f_exp = interpolate.interp1d(exp_data[:,0], exp_data[:,1])
#x_exp = np.linspace(0,30,100)
y_interpolated = f_th(exp_data[:,0])
model_inter = np.vstack((exp_data[:,0],y_interpolated))
print(model_inter)
def diff(x):
    a, residuals = np.linalg.lstsq(exp_data,model_inter*x)
    print(a)
    return a
optimize.minimize_scalar(diff)
fig = plt.figure()
ax  = fig.add_subplot(111)
ax.plot(th_data[:,0], th_data[:,1], color='r')
ax.plot(exp_data[:,0],y_interpolated, 'o',color='b')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

exp_data = np.load("I_q_IPA_exp.npy")
th_data = np.load("I_q_IPA_model.npy")
av1 = np.nanmean(exp_data[:,1])
av2 = np.nanmean(th_data[:,1])
scale = av1/av2
th_data[:,1] = th_data[:,1] * scale
fig = plt.figure()
ax  = fig.add_subplot(111)
ax.plot(exp_data[:,0], exp_data[:,1], color='b')
ax.plot(th_data[:,0], th_data[:,1], color='r')
plt.show()
```
