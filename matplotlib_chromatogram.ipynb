```python
import matplotlib.pyplot as plt

def parse_tab_txt_2list(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    # Split lines into columns
    columns = [line.strip().split('\t') for line in lines]

    # Swap the columns and rows
    rows = list(map(list, zip(*columns)))
    # Convert elements to floats
    n=0
    m=0
    for n in range(0,len(rows)):
        for m in range(0,len(rows[n])):
            try:
                rows[n][m] = float(rows[n][m])
            except ValueError:
                continue
            m+=1
        print(n)
    return rows

#Get data 1
abs280_1 = parse_tab_txt_2list("abs280.txt")
#Get data 2
abs280_2 = parse_tab_txt_2list("abs280_2.txt")
fig = plt.figure()
ax  = fig.add_subplot(111)
#Plot first data
ax.plot(abs280_1[0],abs280_1[1], label="data1", color='b')

#Plot second data
ax.plot(abs280_2[0],abs280_2[1], label="data2", color='g')

ax.set_xlabel(r'Volume (ml)')
ax.set_ylabel(r'Absorbance (mAbs)')
ax.legend(loc=1, frameon=False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('Chromatogram')
plt.show()
```
